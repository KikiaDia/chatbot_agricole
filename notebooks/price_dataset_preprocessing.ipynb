```python
import pandas as pd
df = pd.read_csv(r"notebooks\data\ObservationData_ghzcyke.csv")
df=df.drop(columns=['prix', 'Unit'])

# df['Date'] = df['Date'].str.slice(0, 4) + '/' + df['Date'].str.slice(4, 6)
df['Date'] = pd.to_datetime(df['Date'], format='%YM%m')
# Convert the datetime to the desired format
df['Date'] = df['Date'].dt.strftime('%m/%Y')

df['produits'] = df['produits'].str.replace('Arachide décortiquée', 'Arachide')
df['produits'] = df['produits'].str.replace('Riz importé brisé parfumé', 'Riz')
df['produits'] = df['produits'].str.replace('Arachide non décortiquée', 'Patate')
df['produits'] = df['produits'].str.replace('Riz importé brisé ordinaire', 'Manioc')
df['produits'] = df['produits'].str.replace('Riz local décortiqué', 'Manioc')

df = df.rename(columns={'Value': 'prix/KG'})
df = df.rename(columns={'Date': 'date'})
df = df.rename(columns={'régions': 'regions'})


one_hot_encoded = pd.get_dummies(df['produits'])
df = pd.concat([df, one_hot_encoded], axis=1)

one_hot_encoded = pd.get_dummies(df['regions'])
df = pd.concat([df, one_hot_encoded], axis=1)


df
```
